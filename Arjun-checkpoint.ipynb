```python
import pandas as pd
from sklearn.model_selection import train_test_split  
from sklearn.feature_extraction.text import TfidfVectorizer  


from sklearn.linear_model import LogisticRegression

from sklearn.metrics import classification_report 
import re      
import joblib  
import string  
```

```python
fake=pd.read_csv("Fake.csv")
true=pd.read_csv("True.csv")
```

```python
fake.head()
```

```python
true.head()
```

```python
fake['class']=0
true['class']=1
```

```python
data=pd.concat([fake,true], axis = 0)
```

```python
data.sample(10)
```

```python
data = data.drop(["title","subject","date"], axis=1)
```

```python
data.reset_index(inplace=True)
```

```python
data.drop(['index'],axis = 1,inplace=True)
```

```python
data.sample(5)
```

```python
def clean_text(text):
    text = text.lower()  
    # Converts all characters in the text to lowercase.

    text = re.sub('\[.*?\]', '', text)  
    # Removes text inside square brackets [like this].

    text = re.sub("\\W", " ", text)  
    # Removes all non-word characters (anything except letters, digits, and underscore).

    text = re.sub("https?://\S+|www\.\S+", "", text)  
    # Removes URLs (http, https, www links).

    text = re.sub("<.*?>+", "", text)  
    # Removes any text inside angle brackets <like this>.

    text = re.sub("[%s]" % re.escape(string.punctuation), "", text)  
    # Removes all punctuation symbols.

    text = re.sub("\n", "", text)  
    # Removes newline characters.

    text = re.sub("\w*\d\w*", "", text)  
    # Removes words containing numbers (like "gpt3", "hello123", etc.).

    return text
```

```python
data["text"]=data["text"].apply(clean_text)
```

```python
x=data["text"]
y=data["class"]
xtrain ,xtest ,ytrain, ytest=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
vectorizer = TfidfVectorizer()
xv_train = vectorizer.fit_transform(xtrain)
xv_test = vectorizer.transform(xtest)


```

```python
lr = LogisticRegression()
lr.fit(xv_train, ytrain)
```

```python
prediction= lr.predict(xv_test)
lr.score(xv_test,ytest)
```

```python
print(classification_report(ytest,prediction))
```

```python
joblib.dump(vectorizer,"vectorizer.jb")
joblib.dump(lr,"lr_model.jb")
```
